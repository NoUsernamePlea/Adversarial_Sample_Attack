```python
import matplotlib.pyplot as plt
import json
import seaborn as sns
import numpy as np
import scienceplots
import pdb
import os
import pandas as pd
import timm

# plt.style.use(['science', 'ieee', 'no-latex'])
# plt.style.use(['science', 'no-latex'])
plt.style.use(['science'])
```

```python
data_dir = './evaluation'
model_names = ['resnet50', 'densenet121', 'wide_resnet50_2', 'vgg19', 'inception_v3', ]
# methods = ['DI-FGSM', 'Poincare', 'MI-FGSM', 'TI-FGSM', 'VMI-FGSM', 'VNI-FGSM', 'SSA', 'SIT', 'BAP']
methods = ['DI-FGSM', 'SSA', 'MI-FGSM', 'TI-FGSM', 'SIT', 'PBAP']
markers = ['o', 'D', 's', '^', '*', 'v', 'h', '<', 'p',  '>', 'X', '2', '1']
colors = ['blue', 'limegreen', 'c', 'm', 'y', 'darkviolet', 'purple', 'orange', 'pink', 'olive', 'red']
```

```python
def extract_items(df, fr, to):    
    frs = df[df[2] == fr]
    tos = frs[frs[3] == to]
    args = tos[0]
    fsucs = tos[4]
    tsucs = tos[5]
    return args, fsucs, tsucs

def extract_items_with_method(df, fr, to, m):
    frs = df[df[2] == fr]
    tos = frs[frs[3] == to]
    ms = tos[tos[1] == m]
    args = ms[0]
    fsucs = ms[4]
    tsucs = ms[5]
    return args, fsucs, tsucs

def replace_model(s):
    s = s.replace('wide_resnet50_2', 'WRN50-2')
    s = s.replace('resnet50', 'RN-50')
    s = s.replace('vgg19', 'VGG-19')
    s = s.replace('inception_v3', 'Inc-v3')
    s = s.replace('densenet121', 'DN-121')
    return s

def replace_method(s):
    if s == 'DI-FGSM':
        return 'DIM'
    elif s == 'Poincare':
        return 'Po-TRIP'
    elif s == 'MI-FGSM':
        return 'MI'
    elif s == 'TI-FGSM':
        return 'TIM'
    elif s == 'VMI-FGSM':
        return 'VMI'
    elif s == 'VNI-FGSM':
        return 'VNI'
    elif s == 'BAP':
        return r'$\textbf{IDAA}$'
    return s
```

# Steps

```python
file_path = os.path.join(data_dir, 'tsteps_evaluation.csv')
df = pd.read_csv(file_path, header=None)
alpha = 0.9
```

```python
fig = plt.figure(figsize=(6, 1), dpi=600)
yticks = [0.0, 20.0, 40.0, 60.0, 80.0, 100.0]
dummy_labels = ['', '', '', '', '' , '']
plt.rcParams['text.usetex'] = True

colormap = plt.cm.nipy_spectral
for i, src_name in enumerate(['resnet50']):
    tgt_models = ['densenet121', 'wide_resnet50_2', 'vgg19', 'inception_v3']
    for j, tgt_name in enumerate(tgt_models):
        ax = plt.subplot(1, len(tgt_models), j+1)
        for k, m in enumerate(methods):
            args, fsucs, tsucs = extract_items_with_method(df, src_name, tgt_name, m)
            if j == 0:
                plt.plot(args.to_numpy(), tsucs.to_numpy() * 100.0, label=replace_method(m), 
                    markersize=1.5, marker=markers[k], clip_on=False, alpha=alpha, color=colors[k])
                
                plt.legend(loc=(1.15, 1.1), fontsize=4, ncol=6, frameon=False, labelspacing=0.3, 
                   columnspacing=1.0, markerscale=0.9)
                
                plt.ylabel('tSuc(\%)', labelpad=0, fontsize=6)
                plt.yticks(yticks, fontsize=4)
            else:
                plt.plot(args.to_numpy(), tsucs.to_numpy() * 100.0, 
                    markersize=1.5, marker=markers[k], clip_on=False, alpha=alpha, color=colors[k])
                
                plt.yticks(yticks, labels=dummy_labels)
            plt.ylim(-0.01, 100.)
            plt.grid(True, axis='y', linestyle='--', which='major')
            plt.xticks(args, fontsize=4, rotation=45)
            plt.xlabel('steps', fontsize=6, labelpad=0)
            plt.title(replace_model(tgt_name), fontsize=4, pad=1.2)
    break
plt.subplots_adjust(wspace=0.05)
plt.savefig('steps_{}.pdf'.format(src_name),
        dpi = 600,
        bbox_inches = 'tight')
plt.show()
```

# Group Size

```python
file_path = os.path.join(data_dir, 'n_evaluation.csv')
df = pd.read_csv(file_path, header=None)
methods = ['SIT', 'BAP']
markers = [ '*', 'v', 'h', '<', 'p',  '>', 'X', '2', '1']
colors = ['y', 'darkviolet', 'purple', 'orange', 'pink', 'olive', 'red']
```

```python
fig = plt.figure(figsize=(3, 1), dpi=600)
yticks = [0.0, 20.0, 40.0, 60.0, 80.0, 100.0]
plt.rcParams['text.usetex'] = True

for i, src_name in enumerate(['resnet50']):
    plt.subplot(1, 2, 1)
    for j, tgt_name in enumerate(['densenet121']):
        for k, m in enumerate(methods):
            args, fsucs, tsucs = extract_items_with_method(df, src_name, tgt_name, m)
            if j == 0:
                plt.plot(args.to_numpy(), fsucs.to_numpy() * 100.0, label=replace_method(m), 
                    markersize=1.5, marker=markers[k], clip_on=False, alpha=alpha, color=colors[k])
            else:
                plt.plot(args.to_numpy(), fsucs.to_numpy() * 100.0, 
                    markersize=1.5, marker=markers[k], clip_on=False, alpha=alpha, color=colors[k])
    plt.ylim(-0.01, 1.)
    plt.grid(True, axis='y', linestyle='--', which='major')
    plt.xticks(args, fontsize=4)
    plt.yticks(yticks, fontsize=4)
    plt.ylabel('fSuc(\%)', labelpad=0, fontsize=6)
    plt.xlabel('group size', fontsize=6, labelpad=0)
    plt.legend(loc=(0.85, 1.05), fontsize=4, ncol=4, frameon=False, labelspacing=0.3, 
                   columnspacing=1.0, markerscale=0.9)
    
    plt.subplot(1, 2, 2)
    for j, tgt_name in enumerate(['densenet121']):
        for k, m in enumerate(methods):
            args, fsucs, tsucs = extract_items_with_method(df, src_name, tgt_name, m)
            plt.plot(args.to_numpy(), tsucs.to_numpy() * 100.0, markersize=1.5, marker=markers[k], 
                     clip_on=False, alpha=alpha, color=colors[k])
    plt.ylim(-0.01, 1.)
    plt.grid(True, axis='y', linestyle='--', which='major')
    plt.xticks(args, fontsize=4)
    plt.yticks(yticks, fontsize=4)
    plt.ylabel('tSuc(\%)', labelpad=0, fontsize=6)
    plt.xlabel('group size', fontsize=6, labelpad=0)
    break
plt.subplots_adjust(wspace=0.35)
plt.savefig('n_{}_{}.pdf'.format(src_name, tgt_name),
        dpi = 600,
        bbox_inches = 'tight')
plt.show()
```
